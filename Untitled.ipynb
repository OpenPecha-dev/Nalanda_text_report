```python
from pathlib import Path
import yaml
import re
```

```python
def to_yaml(dict_):
    return yaml.safe_dump(dict_, sort_keys=False, allow_unicode=True,)
```

```python
karchak_text = Path('./data/karchak.txt').read_text(encoding='utf-8')
```

```python
karchak_lines = karchak_text.splitlines()
```

```python
text_id_title = {}
for karchak_line in karchak_lines[1:]:
    karchak_parts = karchak_line.split(',')
    title = karchak_parts[6]
    if not title:
        title = karchak_parts[2]
        if re.search('\d+-\d+_.+?\.txt', title):
            title = re.search('\d+-\d+_(.+?)\.txt', title).group(1)
    text_id_title[karchak_parts[1]] = title
```

```python
text_id_title
```

```python
Path('./data/text_id_titles.yml').write_text(to_yaml(text_id_title), encoding='utf-8')
```

```python
def from_yaml(yml_path):
    return yaml.safe_load(yml_path.read_text(encoding="utf-8"))
```

```python
text_id_titles = from_yaml(Path('./data/text_id_titles.yml'))
```

```python
import collections
```

```python
od = collections.OrderedDict(sorted(text_id_titles.items()))
```

```python
text_id = dict(od)
```

```python
Path('./data/text_id_titles.yml').write_text(to_yaml(text_id), encoding='utf-8')
```
